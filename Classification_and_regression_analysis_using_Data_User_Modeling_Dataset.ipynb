```python
#importing necessary libraries
import pandas as pd
import numpy as np

#importing library for importing data
from google.colab import files
uploaded = files.upload()

#by using pandas library, reading the dataset 
df = pd.read_csv("Data_User_Modeling_Dataset.csv")
df.info()
```

```python
# loading more libraries for Classification analysis
#for decision tree classification
from sklearn.tree import DecisionTreeClassifier 
# for spilitting data in training and testing
from sklearn.model_selection import train_test_split 
# for accuracy of the model
from sklearn import metrics 
```

```python
df.head(20)
```

```python
#Checking the unique values in "UNS" Column
df['UNS'].unique()
```

# DATA Preprocessing

```python
#Determinining the UNS variable as a categorical variable
df2 = df.copy()
df2["UNS"] = df2["UNS"].astype('category')
```

```python
df2.info()
```

```python
#by using label encoding converting categorical values into binary number (AS THERE ARE ONLY 2 LABELS SO, WE CAN USE LABEL ENCODING)
df2["UNS"] = df2["UNS"].cat.codes
df2.head()
```

# Descriptive statistics

```python
# descriptive statistics of these variables by using describe function
df2.describe()
```

# Classification analysis

```python
# Assign the values of the UNS variable to the variable “labels” 
y = df2.UNS
y
```

```python
# dropping "UNS" variable from the data frame.
x = df2.drop(['UNS'], axis= 1)
x
```

## Split dataset as 80% training and 20% test sets.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

## Decision tree algorithm.

```python
# decision tree classifier
dt_clf = DecisionTreeClassifier()

# Fitting the training set
dt_clf = dt_clf.fit(X_train,y_train)

# predit the result from testing set
y_pred_dt = dt_clf.predict(X_test)
```

```python
print("Accuracy of the Decision Tree model:", round(metrics.accuracy_score(y_test, y_pred_dt),2))
```

## Evaluation using accuracy, precision, recall, and F1 score metrics.

```python
# Accuracy of the DT classification model
print("Accuracy of the Decision Tree model:", round(metrics.accuracy_score(y_test, y_pred_dt),2))
```

```python
# importing the library for precision, recall and f1-score metrics.
from sklearn.metrics import classification_report
DecisionTree_report = classification_report(y_test,  y_pred_dt)
print(DecisionTree_report)
```

# Regression analysis

```python
df2
```

```python
#Assigning the values of the PEG variable to the variable “labels” 
y2 = df2.PEG
#dropping "PEG" variable from the data frame.
x2 = df2.drop(['PEG'], axis=1)
y2
```

```python
x2
```

```python
#spiiltting new dataset as 80% training and 20% test sets. 
X2_train, X2_test, y2_train, y2_test = train_test_split(x2, y2, test_size=0.20, random_state=42)
```

## linear multiple regression

```python
#Importing sklearn library for linear regression
from sklearn.linear_model import LinearRegression
ml_regression = LinearRegression() 
#fitting training set in the linear regression model 
ml_regression.fit(X2_train, y2_train)
```

```python
#Intercept and Coefficient for the multilinear equation
print("Intercept: ", ml_regression.intercept_)
print("Coefficients:")
list(zip(x2, ml_regression.coef_))
```

ML Regression Equation: Equation = 0.7029 + (0.03836 * STG) + (0.033996* SCG) + (-0.0128 * STR) + (-0.15654 * LPR) + (-0.44217 * UNS) + e

```python
# Prediction of result from the test set
y_pred_mlr= ml_regression.predict(X2_test)
```

## Evaluation of model by presenting MSE, RMSE, and R-square values.

```python
# Evaluation of the linear regression model
MSE_mlr = round(metrics.mean_squared_error(y2_test, y_pred_mlr),3)
RMSE_mlr = round(np.sqrt(metrics.mean_squared_error(y2_test, y_pred_mlr)),3)
MAE_mlr = round(metrics.mean_absolute_error(y2_test, y_pred_mlr),3)
print('MSE (Mean Square Error):', MSE_mlr)
print('RMSE (Root Mean Square Error):', RMSE_mlr)

print("R-square value:", round(metrics.r2_score(y2_test, y_pred_mlr),3))

print('MAE (Mean Absolute Error):', MAE_mlr)
```
